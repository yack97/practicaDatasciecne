```python
import  pandas as pd
import  numpy as np
```

```python
bikes = pd.read_csv('bikes.csv', index_col='datetime', parse_dates=True)
```

```python
bikes.rename(columns={'count':'total'}, inplace=True)
```

```python
bikes.head()
```

```python
bikes['hora']=bikes.index.hour
```

```python
bikes_dumies = pd.get_dummies(bikes.index.hour, drop_first=True, dtype=int)
```

```python
bikes_dumies.head()
```

```python
bikes_dumies.index = bikes.index
```

```python
bikes_dumies.sample(5)
```

```python
a = np.array(((bikes.index.hour)>=7)&(bikes.index.hour<=20))
```

```python
bikes['dia']= a.astype(int)
bikes['dia'].sample(10)
```

```python
bikes = pd.concat([bikes, bikes_dumies], axis=1)
```

```python
bikes.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import  metrics
```

```python
from  sklearn.model_selection import train_test_split
```

```python
def train_test_rmse (feature_cols):
    X =bikes[feature_cols]
    y=bikes.total
    X_train, X_test,y_train, y_test = train_test_split(X,y,shuffle=False)
    linreg=LinearRegression()
    linreg.fit(X_train,y_train)
    y_pred=linreg.predict(X_test)
    return np.sqrt(metrics.mean_squared_error(y_test, y_pred))
```

```python
train_test_rmse(['temp', 'season', 'humidity' ])
```

```python
train_test_rmse(['temp','season', 'humidity', 'dia'])
```

```python
train_test_rmse(['temp', 'season', 'humidity', 'hora' ])
```

```python
bikes['temp_2']=bikes.temp**2
bikes['humidity_2']=bikes.humidity**2
```

```python
train_test_rmse(['temp', 'season', 'humidity', 'temp_2', 'humidity_2'])
```
